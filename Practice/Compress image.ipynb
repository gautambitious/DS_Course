```python
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image as img
```

```python
image = np.array(img.open("panda.jpg"))
```

```python
def show(arr):
    plt.imshow(arr)
```

```python
show(image)
```

```python
def compress(arr, k=4):
    i,j = 0,0
    result = np.ones((arr.shape[0]//k, arr.shape[1]//k, 3), dtype=np.uint8)
    while i<arr.shape[0]-k:
        while j<arr.shape[1]-k:
            cut = arr[i:i+k, j:j+k]
            mean = np.mean(np.mean(cut, axis=0), axis=0).astype(int)
            result[i//k,j//k] = mean
            j+=k
        i+=k
        j=0
    return result
```

```python
image2 = compress(image, 2)
```
