```python
import numpy as np
import requests 
import json
import pandas as pd
```

```python
target = pd.read_csv('ID.csv')
```

```python
ids = target["ID"].values
```

```python
url = "http://api.icndb.com/jokes/{}"
result = {}
for i in ids:
    joke = requests.get(url.format(i)).json()['value']['joke']
    result[i] = joke
    print("done", i)
```

```python
df = pd.DataFrame(zip(result.keys(), result.values()), columns=["ID", "Joke"])
```

```python
with open("result.csv","w", newline='') as f:
    df.to_csv(f, index=False)
```
