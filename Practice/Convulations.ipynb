```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../datasets/panda.jpg')
```

```python
img = cv2.resize(img, (100,100))
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def show(img, label="image"):
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.title(label+str(img.shape))
    plt.style.use('seaborn')
    plt.show()
```

```python
show(img)
```

```python
img_gray = np.mean(img, axis=2)
```

```python
show(img_gray)
```
