```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../datasets/panda.jpg')
```

```python
img = cv2.resize(img, (100,100))
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def show(img, label="image"):
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.title(label+str(img.shape))
    plt.style.use('seaborn')
    plt.show()
```

```python
show(img)
```

```python
img_gray = np.mean(img, axis=2)
```

```python
show(img_gray)
```

```python
def apply(img, filterArr):
    F = filterArr.shape[0]
    result = np.ones((img.shape[0]-F+1, img.shape[1]-F+1))
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            cut = img[i:i+F, j:j+F]*filterArr
            result[i,j] = abs(np.mean(cut))
    return result
```

```python
filterArr = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
show(apply(img_gray, filterArr))
```
