```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
pil_img = Image.open("doctor-strange.jpg")
```

```python
img = np.array(pil_img)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
X = img.reshape(-1, 3)
```

```python
model = KMeans(10)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
pallets
```

```python
plt.imshow(pallets.reshape(1, -1, 3))
```

```python
gen = np.array([pallets[item] for item in model.labels_])
```

```python
gen.shape
```

```python
plt.imshow(gen.reshape(img.shape))
```
