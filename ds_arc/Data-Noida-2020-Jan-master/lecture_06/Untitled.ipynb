```python
k = 10

tar = np.zeros((img.shape[0]//k, img.shape[1]//k, 3), dtype=np.uint8)

for i in range(0, img.shape[0]-k+1, k):
    for j in range(0, img.shape[1]-k+1, k):
        cut = img[i:i+k, j:j+k]
        tar[i//k, j//k] = np.mean(np.mean(cut, axis=0), axis=0)
        
plt.imshow(tar)
```

```python
k = 30

tar = np.zeros_like(img, dtype=np.uint8)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        cut = img[max(0, i-k//2):i+k//2, max(0, j-k//2):j+k//2]
        tar[i, j] = np.mean(np.mean(cut, axis=0), axis=0)
        
plt.imshow(tar)
```
