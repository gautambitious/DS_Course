```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
```

```python
def load_np_img(location):
    pil_img = Image.open(location)
    img = np.array(pil_img)
    return img
```

```python
cb = load_np_img("cb_tshirt_logo.png")
```

```python
cb = cb[::2, ::2]
```

```python
panda = load_np_img("panda.jpg")
```

```python
plt.imshow(panda)
```

```python
x, y, w, h = 500, 400, cb.shape[1], cb.shape[0]

cut = panda[max(0, y):min(y+h, panda.shape[0]), max(0,x):min(x+w, panda.shape[1])]
```

```python
plt.imshow(cut)
```

```python
for i in range(cut.shape[0]):
    for j in range(cut.shape[1]):
        if cb[i, j, 3] > 100:
            cut[i, j] = cb[i, j, :3]
```
