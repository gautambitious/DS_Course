```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3)
plt.plot(x, x**2)
plt.subplot(2, 2, 4)
plt.plot(x, x**3)
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 2, 4)
plt.plot(x, x**2)
plt.show()
```

```python
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

```python
plt.plot(x, x**2, label="sqare", color="g", marker="^")
plt.plot(x, x**3, label="cube", color="b")
plt.xlabel("Ye hai x ka")
plt.ylabel("Ye hai y ka")
plt.title("Poore 20 number")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.legend()
```

```python
plt.scatter(x, x**2, label="sqare", color="g", marker="^", s=40)
plt.scatter(x, x**3, label="cube", color="b", marker="^")
plt.xlabel("Ye hai x ka")
plt.ylabel("Ye hai y ka")
plt.title("Poore 20 number")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.legend()
```
