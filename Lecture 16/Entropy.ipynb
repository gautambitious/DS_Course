```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balls = 50

entropies = []

for red in range(51):
    blue = balls - red
    bag = ["red"] * red + ["blue"]*blue
    unique, counts = np.unique(bag, return_counts=True)
    prob = counts/np.sum(counts)
    entropy = -np.sum(prob*np.log(prob))
    entropies.append(entropy)
```

```python
plt.plot(entropies)
```
