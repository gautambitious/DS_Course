```python
with open('speech.txt','r',encoding='utf-8') as f:
    text = f.read()
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import numpy as np
```

```python
import string
```

```python
words = word_tokenize(text.lower())
```

```python
bad_words = stopwords.words('English')
```

```python
punc = list(string.punctuation)
```

```python
bad = bad_words+punc
```

```python
pure = [word for word in words if word not in bad]
```

```python
items, counts = np.unique(pure, return_counts=True)
```

```python
last_indices = counts.argsort()[-1000:]
```

```python
bag = items[last_indices]
```

```python
vocab = dict(zip(bag, range(len(bag))))
```

```python
len(vocab)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
model = CountVectorizer(max_features=1000)
```

```python
model.fit(sent_tokenize(text.lower()))
```

```python
sorted(model.vocabulary_.keys())
```
