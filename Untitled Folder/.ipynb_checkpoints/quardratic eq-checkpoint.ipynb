```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0, 100, 100)
a = 5
b = 3
c = 2
y = (a*X*X + b*X + c) + np.random.randint(-5000, 5000, 100)
```

```python
plt.scatter(X, y)
```

```python
class CustomLinearReg:
    
    def __init__(self, lr=0.1, max_iters=10):
        self.lr = lr
        self.max_iters = max_iters
        
    def fit(self, X, y):
        
        self.__X = X
        self.__y = y
        
        self.sq_coef_ = np.random.random()
        self.x_coef_ = np.random.random()
        self.intercept_ = np.random.random()
        
        for _ in range(self.max_iters):
            self.gradient_decend()
        
        self.y_predict_ = self.sq_coef_*X*X + self.x_coef_*X + self.intercept_
    def gradient_decend(self):
        
        delta_sqcoef, delta_xcoef, delta_inter = self.gradient()
        print(delta_sqcoef, delta_xcoef, delta_inter)
        self.sq_coef_ += delta_sqcoef
        self.x_coef_ += delta_xcoef
        self.intercept_ += delta_inter
        
    def gradient(self):
        X, y = self.__X, self.__y
        y_predict = self.sq_coef_*X*X + self.x_coef_*X + self.intercept_
        dif = (y_predict - y)
        z = dif*X
        return -2*self.lr*(np.mean(z*X)), -2*self.lr*(np.mean(z)), -2*self.lr*(np.mean(dif))
```

```python
model = CustomLinearReg(0.00000001, 100)
```

```python
model.fit(X, y)
```

```python
plt.scatter(X, model.y_predict_)
plt.scatter(X, y)
```
