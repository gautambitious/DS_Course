```python
class CustomLinearReg:
    
    def __init__(self, lr=0.1, max_iters=10):
        self.lr = lr
        self.max_iters = max_iters
        
    def fit(self, X, y):
        
        self.__X = X
        self.__y = y
        
        self.coef_ = np.random.random()
        for _ in range(self.max_iters):
            self.gradient_decend()
        self.y_predict_ = self.coef_*(X*X)
    def gradient_decend(self):
        
        delta_coef = self.gradient()
        print(delta_coef)
        
        self.coef_ += delta_coef
        
    def gradient(self):
        X, y = self.__X, self.__y
        y_predict = self.coef_*X*X
        z = (y_predict - y)
        return -(2*np.mean((y_predict - y)*X*X)*self.lr)
```

```python
model = CustomLinearReg(lr=0.00000001, max_iters=100)
model.fit(X, y)
```
