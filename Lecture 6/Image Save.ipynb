```python
from PIL import Image as img
import numpy as np
import matplotlib.pyplot as plt

pil_img = img.open("panda.jpg")

image = np.array(pil_img)
```

```python
def show(arr):
    plt.imshow(arr)
```

```python
show(image)
```

```python
image[:,:,2] = 0
```

```python
show(image)
```

```python
with open("nogreen.jpg", "wb") as f:
    target = img.fromarray(image)
    target.save(f)
```

```python
show(image[::2, ::2])
```

```python
k = 5
```

```python
new = np.zeros((np.shape(image)[0]//k, np.shape(image)[1]//k, 3), dtype= np.uint8)    
```

```python
for i in range(0, image.shape[0]-k,k):
    for j in range(0, image.shape[1]-k,k):
        val=np.mean(np.mean(image[i:i+k, j:j+k], axis=0), axis=0)
        new[i//k,j//k]=val
        
```

```python
show(new)
```
