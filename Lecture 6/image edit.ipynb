```python
from PIL import Image as img
import numpy as np

import matplotlib.pyplot as plt

pil_img = img.open("panda.jpg")

image = np.array(pil_img)

plt.imshow(image)

image[:,:,1] = 0

def show(arr):
    plt.imshow(arr)

show(image)
```

```python
tleft = image[:100, :100]
tright = image[:100, -100:]
bleft = image[-100: , :100]
bright = image[-100: , -100: ]
```

```python
hs = np.hstack([tleft, tright, bleft, bright])
vs = np.vstack([tleft, tright, bleft, bright])
```

```python
show(hs)
```
