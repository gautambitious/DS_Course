```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from PIL import Image as img
```

```python
image = np.array(img.open("./image.jpeg"))
```

```python
image.shape
```

```python
def show(arr):
    plt.imshow(arr)
```

```python
show(image)
```

```python
model = KMeans(8)
```

```python
model.fit(image.reshape(-1,3))
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
show(pallets.reshape(-1,1,3))
```

```python
gen = np.array([pallets[item] for item in model.labels_])
```

```python
show(gen.reshape(image.shape))
```
