```python
from sklearn.linear_model import LogisticRegression as lr
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y= make_blobs(centers=6,random_state=42, center_box=(-20, 20))
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
model = lr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
model.fit(X, y)
```

```python
x1 = np.linspace(-10, 8, 30)
plt.scatter(X[:,0], X[:,1], c=y)
for ((m1, m2), c) in zip(model.coef_, model.intercept_):
    x2 = -(m1*x1 + c)/m2
    plt.plot(x1, x2)
```

```python
color = ["red", "green", "black", "blue", "yellow", "orange"]
yp = model.predict(X)

yp_color = [color[item] for item in yp]
plt.scatter(X[:,0], X[:,1], c=yp_color)
for i, ((m1, m2), c) in enumerate(zip(model.coef_, model.intercept_)):
    x2 = -(m1*x1 + c)/m2
    plt.plot(x1, x2, color=color[i])
plt.ylim(-20,40)
```
