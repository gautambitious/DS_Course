```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(x, y):
    plt.plot(x,y)
```

```python
x = np.linspace(-10, 10, 10000)
y = np.sin(x)
```

```python
plot(x,y)
```

```python
plot(x, x**2)
```

```python
plt.figure()
plt.subplot(2,2,1)
plot(x,y)
plt.subplot(2,2,2)
plot(x, np.cos(x))
plt.subplot(2,2,3)
plot(x, np.tan(x))
plt.subplot(2,2,4)
plot(x, x**2)
```

```python
plt.figure()
plt.subplot(1,2,1)
plot(x, x**2)
plt.subplot(2,2,2)
plot(x, y)
plt.subplot(2,2,4)
plot(x, np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(1,3,1)
plot(x, x**2)
plt.subplot(2,2,2)
plot(x, y)
plt.subplot(2,2,4)
plot(x, np.cos(x))
```

```python
plt.plot(x, x**2, label = "square", color='green')
plt.plot(x, x**3, label = "cube", color='red')
plt.xlabel("damn bruh looks like this is x axis")
plt.ylabel("this one is y axis")
plt.title("damn bruhhhhhh")
plt.legend()
# plt.xlim(-1,1)
# plt.ylim(-1,1)
```

```python
plt.scatter(x, x**2, label = "square", color='green', marker='*', s=2)
plt.scatter(x, x**3, label = "cube", color='red',  marker='*', s=2)
plt.xlabel("damn bruh looks like this is x axis")
plt.ylabel("this one is y axis")
plt.title("damn bruhhhhhh")
plt.legend()
# plt.xlim(-1,1)
# plt.ylim(-1,1)
```
