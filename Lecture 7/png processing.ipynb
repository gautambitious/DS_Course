```python
import numpy as np
from PIL import Image as img
import matplotlib.pyplot as plt
```

```python
def show(arr):
    plt.imshow(arr)
def load(loc):
    return np.array(img.open(loc))
```

```python
logo = load("cblogo.png")
panda = load("panda.jpg")
```
